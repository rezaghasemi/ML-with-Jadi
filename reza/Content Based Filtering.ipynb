# Content Based Filtering

In this method we will provide a recommendor system based on the content of the movies.


## importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## import the data as a dataframe via pandas

```python
df_movies = pd.read_csv('movies.csv')
df_ratings = pd.read_csv("ratings_sample.csv")
```

```python
## Data assessment

df_movies.head()
```

```python
df_ratings.head()
```

## Data Sanitization

```python
## Removing the year in the front of the title of the movies

df_movies['Year'] = df_movies.title.str.extract('(\d\d\d\d)', expand = False)

df_movies.title = df_movies.title.str.replace('(\(\d\d\d\d\))', '')
```

```python
## removing spaces from start and trail

df_movies['title'] = df_movies['title'].apply(lambda x: x.strip())
```

```python
## Spliting the genres

df_movies.genres = df_movies.genres.str.split('|')
```

```python
moviesWithGenres_df = df_movies.copy()
```

```python
moviesWithGenres_df.head()
```

```python
## make new columns and seperate the genres

for (index, row) in moviesWithGenres_df.iterrows():
    for item in row['genres']:
        moviesWithGenres_df.at[index, item] = 1
        
```

```python
moviesWithGenres_df = moviesWithGenres_df.drop('genres',axis = 1)


```

```python
moviesWithGenres_df = moviesWithGenres_df.fillna(0)

moviesWithGenres_df.head()
```

```python
df_ratings = df_ratings.drop('timestamp', axis = 1)

df_ratings.head()
```
