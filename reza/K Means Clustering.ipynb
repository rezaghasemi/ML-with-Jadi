# K-Means Clustering Algorithm
## Importing required Libs

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Generate Data

```python
X,y = make_blobs(n_samples = 5000, centers = ([4, 4], [-2, -1], [2, -3],[1,1]), cluster_std = 0.9 )
```

```python
## Plot the values

print(X.shape)
print(y.shape)

plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("X Values")
plt.ylabel("Y Values")
```

## Loading the Gun :D

```python
k_means = KMeans(init = "k-means++", n_clusters = 4, n_init = 12)
```

## Shooting


```python
k_means.fit(X)
```

```python
X_predict = k_means.predict(X)
```

## ploting the clusters

```python
plt.scatter(X[:, 0], X[:, 1],c = X_predict, marker = '.')
plt.xlabel("X Values")
plt.ylabel("Y Values")
```
